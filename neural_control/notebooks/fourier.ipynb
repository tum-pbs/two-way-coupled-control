```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python

p = 1.6
dt = 0.1
Nmax = int( (p /  dt / 2.))
N = int(0.2 * Nmax)
print(Nmax, N)
max_amplitude = 5.0
max_individual_amplitude = max_amplitude / (0.5 + N)
randomGenerator = np.random.RandomState()
#randomGenerator.seed(0)
amplitudes = randomGenerator.uniform(- max_individual_amplitude, max_individual_amplitude, N)
#randomGenerator.seed(1)
phases = randomGenerator.uniform(- 0.0, 2.0*np.pi, N)

def fourier_series(amplitudes, phases, N, p, x):
    y = amplitudes[0] / 2. 
    for amplitude, phase, n in zip(amplitudes, phases, range(N)): 
        y += amplitude * np.cos(2*np.pi*(n+1)/p * x - phase)
    return y

t = np.arange(0.0, p + dt, dt)
y = np.zeros_like(t)
for i, value in enumerate(t): 
    y[i] = fourier_series(amplitudes, phases, N, p, value) 
plt.plot(t,y)

t = np.linspace(0.0, p, 101 )
y = np.zeros_like(t)
for i, value in enumerate(t): 
    y[i] = fourier_series(amplitudes, phases, N, p, value) 
plt.plot(t+p,y)
```
